```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def genresconvert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
genresconvert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(genresconvert)
```

```python
movies.head(2)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(genresconvert)
```

```python
movies.head(2)
```

```python
movies.iloc[0].cast
```

```python
def castconvert(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter<4:
         l.append(i['name'])
        counter=counter+1
    return l
```

```python
castconvert('[{"cast_id": 242, "character": "Jake Sully", "credit_id": "5602a8a7c3a3685532001c9a", "gender": 2, "id": 65731, "name": "Sam Worthington", "order": 0}]')
```

```python
movies['cast']=movies['cast'].apply(castconvert)
```

```python
movies.head(4)
```

```python
movies.iloc[0].crew
```

```python
def director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
         l.append(i['name'])
       
    return l
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head()
```

```python
def collapse(obj):
    l=[]
    for i in obj:
        l.append(i.replace("",""))
    return l
```

```python
movies['cast']=movies['cast'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['kewords']=movies['keywords'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy Science Fiction culture clash future space war space colony society space travel futuristic romance space alien tribe alien planet cgi marine soldier battle love affair anti war power relations mind and soul 3d Sam Worthington Zoe Saldana Sigourney Weaver Stephen Lang James Cameron.')
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Initialize the CountVectorizer
cv= CountVectorizer()
```

```python
vector=cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distance=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(new.iloc[i[0]]['title'])
```
